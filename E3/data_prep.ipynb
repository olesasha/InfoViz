```python
#nvm
import json
import os

# Define an empty list to store all features
all_features = []

# Define the directory containing the JSON files
directory = 'countries/'

# Loop through all JSON files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.json'):
        country_name = os.path.splitext(filename)[0]  # Extract country name from filename
        with open(os.path.join(directory, filename), 'r') as f:
            data = json.load(f)
            # Append features from each file to the list
            for feature in data['features']:
                feature['properties'] = {'name': country_name}
            all_features.extend(data['features'])

# Create a dictionary with the desired structure
merged_data = {
    "type": "FeatureCollection",
    "features": all_features
}

# Write the merged data to a single JSON file
with open('merged.json', 'w') as f:
    json.dump(merged_data, f)
```
