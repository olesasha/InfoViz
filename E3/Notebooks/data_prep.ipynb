```python
def func(n):
    if n==1: 
        return 1
    if n == 2:
        return 1
    if n % 2 != 0:
        return abs(n*func(n-1))-func(n-2)**2
    if n % 2 == 0:
        return abs(n*func(n-2))-func(n-1)**2
    else: 
        return None
```
