```python
import polars as pl

# Laden der Daten
df_race_data = pl.scan_parquet("./static/data/race_data.parquet")
df_driver_data = pl.scan_parquet("./static/data/all_driver_data.parquet")
df_lap_data = pl.scan_parquet("./static/data/all_laps.parquet")

# Filtern der Daten
df_lap_data = df_lap_data.filter((pl.col("year") == year) & (pl.col("round_number") == round_number)).collect()
df_race_data = df_race_data.filter((pl.col("year") == year) & (pl.col("round_number") == round_number)).collect()
df_driver_data = df_driver_data.filter((pl.col("year") == year) & (pl.col("round_number") == round_number)).collect()

# Auswählen der benötigten Spalten
df_driver_data = df_driver_data.select(['round_number', 'year', 'DriverNumber', 'Abbreviation', 'TeamName', 'TeamColor', 'TeamId'])
df_lap_data = df_lap_data.select(['round_number', 'year', 'DriverNumber', 'LapNumber', 'Position'])

# Überprüfen auf Duplikate vor dem Join
print("Duplikate in df_race_data vor dem Join:", df_race_data.duplicated().sum())
print("Duplikate in df_driver_data vor dem Join:", df_driver_data.duplicated().sum())
print("Duplikate in df_lap_data vor dem Join:", df_lap_data.duplicated().sum())

# Join der DataFrames
df_tester = df_race_data.join(df_driver_data, left_on=['round_number', 'year', 'driver_number'], right_on=['round_number', 'year', 'DriverNumber'])

# Überprüfen auf Duplikate nach dem ersten Join
print("Duplikate in df_tester nach dem ersten Join:", df_tester.duplicated().sum())

# Weiterer Join
df_tester_2 = df_tester.join(df_lap_data, left_on=['round_number', 'year', 'driver_number', 'LapNumber'], right_on=['round_number', 'year', 'DriverNumber', 'LapNumber'])

# Überprüfen auf Duplikate nach dem zweiten Join
print("Duplikate in df_tester_2 nach dem zweiten Join:", df_tester_2.duplicated().sum())
```
